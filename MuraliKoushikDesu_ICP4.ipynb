```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

# 2. Basic statistical description
print(df.describe())

# 3. Check null values
print(df.isnull().sum())

# 3a. Replace null values
df['Calories'].fillna(df['Calories'].mean(), inplace=True)

# 4. Aggregate data
print(df[['Duration','Calories']].agg(['min', 'max', 'count', 'mean']))

# 5. Filter calories 500-1000
calories_filter = df['Calories'].between(500, 1000)
print(df[calories_filter])

# 6. Filter calories > 500 and pulse < 100
filters = (df['Calories'] > 500) & (df['Pulse'] < 100)
print(df[filters])

# 7. New dataframe without Maxpulse
df_modified = df.drop(columns=['Maxpulse'])

# 8. Delete Maxpulse column
del df['Maxpulse']

# 9. Convert Calories to int
df['Calories'] = df['Calories'].astype(int)

# 10. Scatter plot
plt.scatter(df['Duration'], df['Calories'])
plt.title('Duration vs Calories')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.show()
```

```python
import matplotlib.pyplot as plt

programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

fig, ax = plt.subplots()
ax.pie(popularity, labels=programming_languages, autopct='%1.1f%%')
ax.set_title('Popularity of Programming Languages')

plt.show()
```
